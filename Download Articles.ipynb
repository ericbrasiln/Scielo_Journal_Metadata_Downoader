# Download Articles from a Journal in SCielo

This notebook presents the process to download all the articles from a specific journal in the SCielo database. Fisrs to use BeatifulSoup to inspect the Journal page in SCielo for the links for each of the issues.  

```python
from bs4 import BeautifulSoup
import requests, time
```

```python
#Given a journal all issues page on SCielo downloads all the articles in XML format

how_journal = "http://www.scielo.org.co/scielo.php?script=sci_issues&pid=0120-5927&lng=en&nrm=iso"
CALJ = "http://www.scielo.org.co/scielo.php?script=sci_issues&pid=0123-4641&lng=en&nrm=iso"
ikala = "http://www.scielo.org.co/scielo.php?script=sci_issues&pid=0123-3432&lng=en&nrm=iso"

journal_url = ""

journal = requests.get(ikala)

soup = BeautifulSoup(journal.content, 'html.parser')

links = soup.find_all("a", href=True)

issues = []

for link in links:
    if "http://www.scielo.org.co/scielo.php?script=sci_issuetoc&pid" in link["href"]:
        issues.append(link["href"])
       
```

After we have the links stored in the list **issues** we can go to each of the links and look for the link to each of the articles. 

```python
#Gets a link for all the html files in each issue

links_to_download = []

def get_article_links(enlace):
    try:        
        iss = requests.get(enlace)  
        time.sleep(1)
        soup = BeautifulSoup(iss.content, 'html.parser')
        links = soup.find_all(["a"])
        for link in links: 
            if "http://www.scielo.org.co/scielo.php?script=sci_arttext&pid" in link["href"]:
                links_to_download.append(link["href"])  
    except KeyError:
        print(enlace + "links Not Found")
        
            
for issue in issues:
    get_article_links(issue)

len(links_to_download)
```

The code returns le number of links that it has stored on the **links_to_download** list. It could be a big number, so the download process could take long as it is necessary to wait 2 seconds between downloads to avoid being rejected by the SCielo server. 

```python
root = "D:/journal_articles/xml_articles/"

def download_xml(link):
    try:
        art = requests.get(link)
        soup = BeautifulSoup(art.content, 'html.parser')
        xml_link = soup.find(target="xml")['href']
        r = requests.get(xml_link)
        sp = BeautifulSoup(r.content, "html.parser")
        name = sp.front.find("article-id").string
        time.sleep(2)
        print(name)
        with open(root + name + ".xml","wb") as f:
            f.write(r.content)
    except TypeError: 
        print(link + " Not Found")


for link in links_to_download:
    download_xml(link)
    
    
```
